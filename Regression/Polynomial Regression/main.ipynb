```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('./Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
plt.scatter(X, y, color='red')
plt.title('Salary based on Job Position')
plt.xlabel('Job Position Level')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
```

```python
y_pred = lin_reg.predict(X_test)
```

```python
plt.scatter(X, y, color='red')
plt.plot(X, lin_reg.predict(X), color='blue')
plt.title('Salary based on Job Position')
plt.xlabel('Job Position Level')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
poly = PolynomialFeatures(degree=4)
poly.fit(X_train)
X_train_poly = poly.transform(X_train)
X_test_poly = poly.transform(X_test)
```

```python
poly_lin_reg = LinearRegression()
poly_lin_reg.fit(X_train_poly, y_train)
```

```python
y_poly_pred = poly_lin_reg.predict(X_test_poly)
```

```python
plt.scatter(X, y, color='red')
plt.plot(X, poly_lin_reg.predict(poly.transform(X)), color='blue')
plt.title('Salary based on Job Position')
plt.xlabel('Job Position Level')
plt.ylabel('Salary')
plt.legend()
plt.show()
```
