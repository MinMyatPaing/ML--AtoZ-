```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
dataset = pd.read_csv('./Mall_Customers.csv')
X = dataset.iloc[:, [3, 4]].values
print(X)
```

```python
sc = StandardScaler()
X = sc.fit_transform(X)
print(X)
```

```python
import os
os.environ["OMP_NUM_THREADS"] = "1"  # Set the number of threads for parallel processing
wcss = []  # List to store the within-cluster sum of squares (WCSS) for each number of clusters
for i in range(1, 11):  # Iterate over the range of cluster numbers
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)  # Initialize the K-Means model
    kmeans.fit(X)  # Fit the model to the data
    wcss.append(kmeans.inertia_)  # Append the WCSS value for the current number of clusters
```

```python
plt.plot(range(1, 11), wcss, color = 'blue')
plt.xlabel('Number of Centroids')
plt.ylabel('wcss')
plt.title('Elbow Method')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X)
print(y_kmeans)
print(kmeans.cluster_centers_)
```

```python
X = sc.inverse_transform(X)  # Transform the scaled data back to its original scale
centroids = sc.inverse_transform(kmeans.cluster_centers_)  # Transform the cluster centroids back to the original scale
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')  # Plot points in Cluster 1
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')  # Plot points in Cluster 2
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')  # Plot points in Cluster 3
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')  # Plot points in Cluster 4
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')  # Plot points in Cluster 5
plt.scatter(centroids[:, 0], centroids[:, 1], s = 300, c = 'yellow', label = 'Centroids')  # Plot the centroids
plt.title('Clusters of customers')  # Set the title of the plot
plt.xlabel('Annual Income (k$)')  # Label the x-axis
plt.ylabel('Spending Score (1-100)')  # Label the y-axis
plt.legend()  # Add a legend to the plot
plt.show()  # Display the plot
```
